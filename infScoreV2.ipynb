```python
import time
import urllib.request
import json
import pandas as pd
import re
import sklearn
from sklearn import preprocessing
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv("hascolUserData.csv")
df
```

```python
df['Reach Score'] = (df['Friends Count'] - df['Users Following Count'] ).abs()
df
```

```python
X =df[['Retweet Count','Users Following Count','Friends Count','Users Number of Likes','Users Tweets Count' , 'Lists Containing User']]
Y = df['Reach Score']
```

```python
X_split = int(0.7 * len(X))
y_split = int(0.7 * len(Y))
X_train = X[: X_split]
X_test = X[X_split:]
y_train = Y[: y_split]
y_test = Y[y_split:]

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#model_Linear_Regression
model = LinearRegression()
model.fit(X_train, y_train.ravel())
df_feature_importance = pd.DataFrame(model.coef_, index=X.columns, columns=['Weights']).sort_values('Weights', ascending=False)
df_feature_importance
```

```python
#model_Decision_tree_regressor
from sklearn.tree import DecisionTreeRegressor
model1 = DecisionTreeRegressor()
model1.fit(X_train, y_train.ravel())
df_feature_importance = pd.DataFrame(model1.feature_importances_, index=X.columns, columns=['Weights']).sort_values('Weights', ascending=False)
df_feature_importance
```
